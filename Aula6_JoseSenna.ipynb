```python
import numpy as np
import matplotlib.pyplot as plt
```

PLOTAGEM DAS EQUAÇÕES DO SISTEMA 1

```python
x1 = np.linspace(-10,10,num=10)
x2 = np.linspace(-10,10,num=10)

plt.title("Equações")
plt.plot(x1, 3 - 2*x1)
plt.plot(x1, (2 + x1)/3)
plt.show()
```

RESOLUÇÃO DO SISTEMA 1

```python
a = np.array([[2,1],[1,-3]])
b = np.array([3,-2])
x = np.linalg.solve(a,b)
print(x)
```
